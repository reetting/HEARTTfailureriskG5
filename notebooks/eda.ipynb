```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
```

```python
# Charger le dataset depuis UCI

heart_failure_clinical_records = fetch_ucirepo(id=519)

X = heart_failure_clinical_records.data.features
y = heart_failure_clinical_records.data.targets

data = pd.concat([X, y], axis=1)

print("Dataset shape:", data.shape)

data.head()
```

```python
# Vérification des valeurs manquantes

missing_values = data.isnull().sum()

print("Missing values per column:")
print(missing_values)

print("\nTotal missing values:", missing_values.sum())
```

### Missing Values Analysis

The dataset was analyzed for missing values using the isnull() function in pandas.
The results show that no missing values are present in any of the variables.

Therefore, no data imputation or cleaning was required.

```python
# Détection des valeurs aberrantes

Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((data < lower_bound) | (data > upper_bound))

print("Number of outliers per feature:")
print(outliers.sum())
```

```python
#Visualisation
plt.figure(figsize=(14,8))
sns.boxplot(data=data)

plt.title("Boxplot of Dataset Features")
plt.xticks(rotation=45)

plt.show()
```

### Outliers Analysis

Outliers were detected using the Interquartile Range (IQR) method and visualized using boxplots.

Some variables such as:

creatinine_phosphokinase

serum_creatinine

platelets

contain several extreme values.

However, in medical datasets these values may represent real pathological conditions, therefore they were not removed. Instead, robust machine learning models will be used to reduce their impact.


```python
print(data["death_event"].value_counts())

print(data["death_event"].value_counts(normalize=True) * 100)
```

```python
#visualisation
sns.countplot(x="death_event", data=data)

plt.title("Class Distribution of Death Event")
plt.xlabel("Death Event")
plt.ylabel("Number of Patients")

plt.show()
```

```python
X = data.drop("death_event", axis=1)
y = data["death_event"]

smote = SMOTE(random_state=42)

X_resampled, y_resampled = smote.fit_resample(X, y)

print("Class distribution after SMOTE:")
print(y_resampled.value_counts())

sns.countplot(x=y_resampled)

plt.title("Balanced Class Distribution after SMOTE")
plt.xlabel("DEATH_EVENT")
plt.ylabel("Number of Patients")

plt.show()
```

### Class Imbalance

he dataset is imbalanced.

Class distribution:

Survived (0): ~68%

Deceased (1): ~32%

This imbalance can negatively affect machine learning models because they may become biased toward the majority class.

To address this issue, the SMOTE (Synthetic Minority Oversampling Technique) method was applied.

SMOTE generates synthetic samples for the minority class, resulting in a balanced training dataset
